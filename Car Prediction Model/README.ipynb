
# Car Price Prediction Project

## Overview

This project aims to predict the selling prices of cars using machine learning techniques. We utilize a dataset containing car-related information to build and evaluate regression models for accurate price predictions. The project covers data collection, preprocessing, model training, and evaluation using Linear Regression, Lasso Regression, and Random Forest Regression.

## Getting Started

Follow these instructions to get the project up and running on your local machine.

### Prerequisites

- Python (>=3.6)
- Jupyter Notebook or any Python IDE
- Required libraries: pandas, matplotlib, seaborn, scikit-learn

To install the necessary libraries, you can use the following command:

```
pip install pandas matplotlib seaborn scikit-learn
```

### Project Structure

The project follows this structure:

- `car_data.csv`: The dataset containing car-related information.
- `car_price_prediction.ipynb`: The Jupyter Notebook with the code for data processing and model training.

## Data Collection and Processing

We begin by loading the car dataset and performing initial data exploration:

- Checking the initial values of the DataFrame
- Checking for missing values
- Checking the distribution of categorical variables

The data preprocessing steps include converting string values to numeric values (encoding) and splitting the data into features and the target variable.

## Model Training and Evaluation

We train and evaluate three regression models:

1. Linear Regression
2. Lasso Regression
3. Random Forest Regression

- Model training using training data
- Model evaluation using R-squared (R^2) score on training and test data
- Visualizing actual vs. predicted prices

## Model Comparison

We compare the models' performance based on the test R^2 scores. The model with the highest test R^2 score is considered the best for this specific dataset.

## Results

Based on the provided R^2 scores:

- Random Forest Regression achieved a test R^2 score of 0.981, indicating the best accuracy.
- Lasso Regression achieved a test R^2 score of 0.871.
- Linear Regression achieved a test R^2 score of 0.837.
